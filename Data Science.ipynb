```python
price_cond = diamonds.price.between(3000, 4000)
is_ideal_cut = diamonds.cut == 'Ideal'
my_diamonds = diamonds[price_cond & is_ideal_cut]
```

```python
my_diamonds
```

```python
my_diamonds.sort_values('carat', ascending=False, inplace=True)
```
