```python
diamonds[diamonds['cut'] == 'Ideal']
```

```python
# This is a "mask" or a "filter"
is_ideal = diamonds['cut'] == 'Ideal'
diamonds[is_ideal]
```

```python
# Diamond between $3,000 and $4,000
# When filtering, use & instead of "and"
diamonds[(diamonds.price >= 3000) & (diamonds.price <= 4000)]
```

```python
# Conditions:
# The heaviest diamond that...
# Between 3k and 4k
# Ideal cut
price_cond = diamonds.price.between(3000, 4000)
is_ideal = diamonds.cut == 'Ideal'
my_diamonds = diamonds[price_cond & is_ideal]
my_diamonds.head()
```

```python
my_diamonds.sort_values('carat', ascending=False, inplace=True)
```

```python
my_diamonds.head(5)
```
